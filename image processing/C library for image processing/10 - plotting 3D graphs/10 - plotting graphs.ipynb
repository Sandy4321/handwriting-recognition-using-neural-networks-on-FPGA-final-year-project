# displaying dynamically refreshing graph

```python
import matplotlib
%matplotlib inline
import time
import pylab as pl
from IPython import display

for i in range(10):
    pl.plot(pl.randn(100))
    display.clear_output(wait=True)
    display.display(pl.gcf())
    time.sleep(0.1)
    
    # used for smoothing out the display process 
    display.clear_output(wait=True)
    display.display(pl.gcf())

display.clear_output()
```
